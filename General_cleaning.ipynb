# MERGING 2 tables - FILMS & OLD_HDD 

### DESCARGA DEL ARCHIVO 1, (film) YA LIMPIO PARA HACER LA FUNCION MERGE

```python
film_joining = pd.read_csv('/Users/JGR/IRON_HACK/w3-project_SQL/archivos_limpios/film_cleaned.csv')
```

### DESCARGA DEL ARCHIVO 2, (old_hdd) YA LIMPIO PARA HACER LA FUNCION MERGE

```python

old_hdd_joining = pd.read_csv('/Users/JGR/IRON_HACK/w3-project_SQL/archivos_limpios/old_hdd_cleaned.csv')
```

### JUNTAMOS LAS DOS TABLAS POR LA COLUMNA DEL TITULO.

```python
merged_films_oldd_hdd = pd.merge(film_joining, old_hdd_joining, on='title')
merged_films_oldd_hdd.head(2)
```

## Limpieza de DATAFRAME

### Borrado de columnas repetidas, (release_year_y).

```python
merged_films_oldd_hdd = merged_films_oldd_hdd.drop(columns=['release_year_y'], axis=1) 
merged_films_oldd_hdd.head(2)
```

### Covertir los minutos del lengh a Horas. 

```python
def conv_min_a_horas(minutos):
    horas = minutos // 60
    minutos_restantes = minutos % 60
    return f"{horas}:{minutos_restantes:02d}"
```

```python
merged_films_oldd_hdd['duration_film'] = merged_films_oldd_hdd['length'].apply(conv_min_a_horas)
```

```python
merged_films_oldd_hdd.head(2)
```

```python
film_and_oldd_hdd = merged_films_oldd_hdd.drop(columns=['length'], axis=1) 
```

```python
film_and_oldd_hdd.head(3)
```

### Envio de data frame finalizada a mi laptop.

```python
film_and_oldd_hdd.to_csv('film_and_oldd_hdd.csv', index=False)
```
